```python
diary_data = json_normalize(data=d['data'], record_path='diary', 
                            meta=['hhcmp', 'service', 'srvy_id', 'diary_id', 'dst_code',
       'smplu_id', 'agreedTac', 'sample_id', 'diary_type', 'luw_switch',
       'rspndnt_id', 'access code'])
```

```python
diary_data.head()
```

```python
diary_data['stop_time'] = pd.to_datetime(diary_data['stop_time'], format='%H:%M')
```

```python
diary_data['start_time'] = pd.to_datetime(diary_data['start_time'], format='%H:%M')
```

```python
diary_data['duration']=diary_data['stop_time']-diary_data['start_time']
```
